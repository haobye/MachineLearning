```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np; 
np.random.seed(0)
import seaborn as sns;
sns.set()
```

```python
uniform_data = np.random.rand(3, 3)
print (uniform_data)
heatmap = sns.heatmap(uniform_data)#例子：特性与特征之间的相关系数
```

```python
ax = sns.heatmap(uniform_data, vmin=0.2, vmax=0.5)#0.2~0.5的取值
ax
```

```python
normal_data = np.random.randn(3, 3)
print (normal_data)
ax = sns.heatmap(normal_data, center=0)
```

```python
import pandas as pd
flights = pd.read_csv("flights.csv")#航班数据
flights.head()
```

```python
flights = flights.pivot("month","year","passengers")
#flights = flights.pivot_table(index="month",values="passengers")
print (flights)
ax = sns.heatmap(flights)
```

```python
ax = sns.heatmap(flights, annot=True,fmt="d")
```

```python
ax = sns.heatmap(flights,annot=True,fmt="d", linewidths=.9)
```

```python
ax = sns.heatmap(flights, cmap="YlGnBu",linewidths=.9)#cmap="YlGnBu"指定色板
```

```python
ax = sns.heatmap(flights, cbar=False)#cbar=False右边的标示没有了
```
